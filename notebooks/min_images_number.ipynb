```python
import os
from src.config import ROOT_DIR

# ============================================================
# ROOTS
# ============================================================

PROCESSED_IMG_DIR = os.path.join(ROOT_DIR, "dataset_processed")

# ============================================================
# BUSCAR MÍNIMO NÚMERO DE XMLs
# ============================================================

min_xml_count = float('inf')

min_folder = None

# Guardar resultados
folder_counts = []

# ============================================================
# RECORRER DATASET
# ============================================================

for root, dirs, files in os.walk(PROCESSED_IMG_DIR):

    # contar archivos .xml
    xml_files = [

        f for f in files

        if f.endswith(".xml")
    ]

    xml_count = len(xml_files)

    # ignorar carpetas vacías
    if xml_count > 0:

        folder_counts.append((root, xml_count))

        print(f"{root} --> {xml_count} XML files")

        # actualizar mínimo
        if xml_count < min_xml_count:

            min_xml_count = xml_count

            min_folder = root

# ============================================================
# RESULTADO FINAL
# ============================================================

print("\n===================================================")

print("MINIMUM XML COUNT")

print("===================================================")

print(f"Folder with minimum XMLs:")

print(min_folder)

print(f"\nMinimum XML count: {min_xml_count}")

# ============================================================
# VALOR PARA BALANCEAR PACIENTES
# ============================================================

MAX_IMAGES_PER_PATIENT = min_xml_count

print(f"\nMAX_IMAGES_PER_PATIENT = {MAX_IMAGES_PER_PATIENT}")
```

```python
import os

from src.config import ROOT_DIR

# ============================================================
# ROOT
# ============================================================

PROCESSED_IMG_DIR = os.path.join(ROOT_DIR, "dataset_processed")

# ============================================================
# BUSCAR MEJOR SESIÓN DE CADA PACIENTE
# ============================================================

patient_best_sessions = []

# cada carpeta de primer nivel = paciente
for patient_id in sorted(os.listdir(PROCESSED_IMG_DIR)):

    patient_path = os.path.join(
        PROCESSED_IMG_DIR,
        patient_id
    )

    if not os.path.isdir(patient_path):
        continue

    best_session = None

    best_count = 0

    # recorrer sesiones del paciente
    for session_id in os.listdir(patient_path):

        session_path = os.path.join(
            patient_path,
            session_id
        )

        if not os.path.isdir(session_path):
            continue

        xml_count = len([
            f
            for f in os.listdir(session_path)
            if f.endswith(".xml")
        ])

        print(
            f"Patient {patient_id} | "
            f"Session {session_id} -> "
            f"{xml_count} XMLs"
        )

        if xml_count > best_count:

            best_count = xml_count

            best_session = session_id

    patient_best_sessions.append({

        "patient_id": patient_id,

        "best_session": best_session,

        "xml_count": best_count
    })

# ============================================================
# RESULTADOS
# ============================================================

print("\n===================================================")
print("BEST SESSION PER PATIENT")
print("===================================================")

for item in patient_best_sessions:

    print(
        f"Patient {item['patient_id']} "
        f"-> Session {item['best_session']} "
        f"({item['xml_count']} images)"
    )

# ============================================================
# MENOR ENTRE LOS MÁXIMOS
# ============================================================

worst_best_patient = min(
    patient_best_sessions,
    key=lambda x: x["xml_count"]
)

print("\n===================================================")
print("MINIMUM OF THE BEST SESSIONS")
print("===================================================")

print(
    f"Patient: {worst_best_patient['patient_id']}"
)

print(
    f"Session: {worst_best_patient['best_session']}"
)

print(
    f"Images : {worst_best_patient['xml_count']}"
)

MAX_IMAGES_PER_PATIENT = worst_best_patient["xml_count"]

print(
    f"\nMAX_IMAGES_PER_PATIENT = "
    f"{MAX_IMAGES_PER_PATIENT}"
)
```
