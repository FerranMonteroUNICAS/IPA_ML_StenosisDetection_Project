```python
import os
from src.config import ROOT_DIR

# ============================================================
# ROOTS
# ============================================================

PROCESSED_IMG_DIR = os.path.join(ROOT_DIR, "dataset_processed")

# ============================================================
# BUSCAR MÍNIMO NÚMERO DE XMLs
# ============================================================

min_xml_count = float('inf')

min_folder = None

# Guardar resultados
folder_counts = []

# ============================================================
# RECORRER DATASET
# ============================================================

for root, dirs, files in os.walk(PROCESSED_IMG_DIR):

    # contar archivos .xml
    xml_files = [

        f for f in files

        if f.endswith(".xml")
    ]

    xml_count = len(xml_files)

    # ignorar carpetas vacías
    if xml_count > 0:

        folder_counts.append((root, xml_count))

        print(f"{root} --> {xml_count} XML files")

        # actualizar mínimo
        if xml_count < min_xml_count:

            min_xml_count = xml_count

            min_folder = root

# ============================================================
# RESULTADO FINAL
# ============================================================

print("\n===================================================")

print("MINIMUM XML COUNT")

print("===================================================")

print(f"Folder with minimum XMLs:")

print(min_folder)

print(f"\nMinimum XML count: {min_xml_count}")

# ============================================================
# VALOR PARA BALANCEAR PACIENTES
# ============================================================

MAX_IMAGES_PER_PATIENT = min_xml_count

print(f"\nMAX_IMAGES_PER_PATIENT = {MAX_IMAGES_PER_PATIENT}")
```
