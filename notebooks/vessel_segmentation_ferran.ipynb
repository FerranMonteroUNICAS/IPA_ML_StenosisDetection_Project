## Ridge filters analysis
Ridge filters are designed to enhance tubular structures like vessels. They work by analyzing the second-order derivatives of the image to identify ridge-like features. The main parameters to tune are:
- **Type of filter**: Frangi, Sato or Meijering.
- **Scale range**: the range of scales at which to analyze the image. This controls the size of vessels that are enhanced.

```python
browse_ridge_grid(series, kernel_size=KERNEL_SIZE, h=H, clip_limit=CLIP_LIMIT, seed=SEED)
```

## Hysteresis Thresholding (Binarization)
This step converts our continuous [0, 1] vesselness maps into a clean binary mask (0 or 255) to isolate the vessels and prepare them for skeletonization.

Using a dual-threshold setup:
- **High Thresh**: Selects the strong, unmistakable core tracks of main arteries.
- **Low Thresh**: Recursively rescues fainter, lower-contrast tracks *only* if they directly connect back to a strong core branch. This bridges vessel discontinuities without magnifying empty background noise.
